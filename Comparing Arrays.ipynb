```python
# Compare sizes of the ICESat and CmCt and extract data missing from CmCt
```

```python
zoom_bounds = [75, 80, 310, 320]
```

```python
h5_files = []
cmct_files = []

for group in h_groups:
    h5_files.extend(group)
    
for group in ct_groups:
    cmct_files.extend(group)
```

```python
# Plots has title, x and y labels, the cmap
# This code goes through the different groups (CmCt and ICESat cpunterpart files and plots them side by side)
# Also shows the data size as well and if there is a size difference between the original and the transferred
    # The H5 data (alt) are larger then the CmCt sizes (from observation)
```

```python
for i in range(len(h5_files)):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    # Lays out the plots properly 
    fig.tight_layout()
    
    h5_data = utils.h5_extract_data(file=h5_files[i], bbox=zoom_bounds)
    cmct_data = utils.cmct_extract_data(cmct_files[i], zoom_bounds)
    
    h5_length_alt = len(h5_data['alt'])
    cmct_length_alt = len(cmct_data['alt'])
    
    print(h5_files[i])
    print(cmct_files[i])
    print(f'Alt Data Size the Same: {h5_length_alt == cmct_length_alt}')
    print(f'Data h5 Points Sizes: {h5_length_alt}')
    print(f'Data cmct Points Sizes: {cmct_length_alt}')
    print(f'Size Difference: {h5_length_alt - cmct_length_alt}')
    print('\n')
    
    utils.plot_h5(h5_data, ax1)
    alt = utils.plot_cmct(cmct_data, ax2)
    fig.suptitle('Difference between geophysical and surface elevation', 
                 y=1.1)
    fig.colorbar(alt)
    
    plt.show()
```
